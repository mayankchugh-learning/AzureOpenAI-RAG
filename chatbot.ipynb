#### Importing Utilities and libraries and also root SSL Certificates

```python
import os
from pathlib import Path

import certifi
from dotenv import load_dotenv
from openai import AzureOpenAI

os.environ["SSL_CERT_FILE"] = certifi.where()

for _p in (Path.cwd() / ".env", Path.cwd().parent / ".env"):
    if _p.is_file():
        load_dotenv(_p)
        break
else:
    load_dotenv()
```

#### Creating an Azure OpenAI Client


```python
_endpoint = (os.environ.get("OPENAI_API_BASE") or "").strip().rstrip("/")
_api_key = os.environ.get("OPENAI_API_KEY", "")
_api_version = os.environ.get("OPENAI_API_VERSION", "2024-02-15-preview")

if not _endpoint or not _api_key:
    raise ValueError("Set OPENAI_API_BASE and OPENAI_API_KEY in your .env file.")

client = AzureOpenAI(
    azure_endpoint=_endpoint,
    api_key=_api_key,
    api_version=_api_version,
)

```

#### Making use of the Chat Completions API

```python
_deployment = os.environ.get("OPENAI_CHAT_DEPLOYMENT") or os.environ.get("RAG_CHAT_DEPLOYMENT")
if not _deployment:
    raise ValueError("Set OPENAI_CHAT_DEPLOYMENT (or RAG_CHAT_DEPLOYMENT) in your .env file.")

response = client.chat.completions.create(
    model=_deployment,
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "List all the players in the Indian national cricket team?"},
    ],
)
print(response.choices[0].message.content)




```
